```python
# =====================================================================
# Control Flow & Basic Logic
# =====================================================================


# 1. Function to check if a number is even or odd
def evaluate_parity(number):
    if number % 2 == 0:
        return f"{number} is Even"
    else:
        return f"{number} is Odd"

# 2. Function to add numbers from 1 up to a limit
def compute_cumulative_sum(limit):
    total = 0
    for i in range(1, limit + 1):
        total = total + i
    return total

# Execution
print("--- Stage 1 Output ---")
print(evaluate_parity(42))
print("Sum from 1 to 50:", compute_cumulative_sum(50))
```

```python
# =====================================================================
# Native Data Structures
# =====================================================================

def analyze_data_structures():
    # List: Keeps items in exact order, allows duplicates
    skills = ["SQL", "Python", "Power BI"]
    skills.append("Scikit-Learn")
    
    # Dictionary: Stores data in Key-Value pairs (like a profile)
    analyst_profile = {"name": "Subhankar", "role": "Data Analyst", "experience": 2}
    
    # Set: Automatically removes any duplicate values
    unique_tags = {"data", "ml", "data", "python"} 
    
    print("List:", skills)
    print("Dictionary:", analyst_profile)
    print("Set (Notice duplicates are gone):", unique_tags)

# Execution
print("--- Stage 2 Output ---")
analyze_data_structures()
```

```python
# =====================================================================
# Functions & Flexible Arguments
# =====================================================================

# Lambda: A quick, one-line function to square numbers
square_numbers = lambda nums: [x**2 for x in nums]

# Flexible function that accepts ANY number of extra inputs
def dynamic_logger(*args, **kwargs):
    print("Positional inputs (*args):", args)
    print("Keyword inputs (**kwargs):", kwargs)

# Execution
print("--- Stage 3 Output ---")
print("Squared Numbers:", square_numbers([1, 2, 3, 4]))
dynamic_logger("Log_01", "Success", user="Subhankar", status="Active")
```

```python
# =====================================================================
# Object-Oriented Programming (OOP)
# =====================================================================

# Parent Class
class DataModel:
    def __init__(self, model_name, framework):
        self.model_name = model_name        # Public (anyone can see/change)
        self.__framework = framework        # Private (hidden using '__')

    # A safe method to look at the hidden variable
    def get_framework(self):
        return self.__framework

# Child Class inherits everything from DataModel
class PredictiveModel(DataModel):
    def train(self):
        return f"Training {self.model_name} using {self.get_framework()}"

# Execution
print("--- Stage 4 Output ---")
my_model = PredictiveModel("RandomForest", "Scikit-Learn")
print(my_model.train())
```

```python
# =====================================================================
# Advanced Python (Decorators & Comprehensions)
# =====================================================================

import time

# A Decorator to measure execution time
def execution_timer(func):
    def wrapper(data_range):
        start_time = time.perf_counter()
        result = func(data_range) # Runs your actual function here
        end_time = time.perf_counter()
        print(f"Function took {end_time - start_time:.6f} seconds to run.")
        return result
    return wrapper

# Using the decorator on our function
@execution_timer
def optimized_data_filter(data_range):
    # List comprehension: filters numbers divisible by 5 in one fast line
    return [x for x in range(data_range) if x % 5 == 0]

# Execution
print("--- Stage 5 Output ---")
filtered_list = optimized_data_filter(50)
print("Filtered Matrix:", filtered_list)
```
